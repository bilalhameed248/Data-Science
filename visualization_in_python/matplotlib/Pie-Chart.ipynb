## Pie Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
classes=["Python", "R", "Machine Learning","Artificial Intelligrence", "Data Science"]
class1_std=[45,15,35,25,30]
```

```python
plt.figure(figsize=(13,7))
plt.pie(class1_std, labels=classes, autopct="%0.2f%%")
plt.show()
```

```python
plt.figure(figsize=(13,7))
explode=[0,0,0.1,0,0]
colors=["r","g", "b","y", "m"]
textprops={"fontsize":15}
plt.pie(class1_std, labels=classes, explode=explode, colors=colors, autopct="%0.2f%%",radius=1.4, textprops=textprops)
plt.legend(loc=3)
plt.show()
```
