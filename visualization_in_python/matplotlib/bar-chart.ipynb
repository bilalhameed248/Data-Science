## Ploting Bar Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
classes=["Python", "R", "AI", "ML", "DS"]

class1_std=[30,10,20,25,10]
class2_std=[40,5,22,20,10]
class3_std=[35,5,24,15,15]
```

## Single Feature Bar Chart

```python
plt.bar(classes, class1_std, width=0.2, color='r' , align='edge', edgecolor='m', linewidth=5, label="Class One Std")
plt.legend(loc=1)
```

```python
plt.barh(classes, class1_std, color='r' , align='edge', edgecolor='m', linewidth=5, label="Class One Std")
plt.legend(loc=1)
```

## Plot Multiple Bar Charts

```python
style.use("ggplot")
plt.figure(figsize=(13,7))
plt.bar(classes, class1_std, width=0.6, color='g', edgecolor='b', linewidth=5, label="Class One Std")
plt.title("Bar Chart of IT Class")
plt.xlabel("Classes")
plt.ylabel("No. of student ")
plt.legend(loc=1)
```

## Plot Multiple Bar Charts Ready

```python
style.use("ggplot")
plt.figure(figsize=(13,7))

classes_index=np.arange(len(classes))
width=0.2
classes_index
plt.bar(classes_index, class1_std, width, color='g', label="Class One Std")
plt.bar(classes_index+width, class3_std, width=0.2, color='b', label="Class Two Std")
plt.xticks(classes_index+width, classes, rotation=30)

plt.title("Bar Chart of IT Class")
plt.xlabel("Classes")
plt.ylabel("No. of student ")
plt.legend(loc=1)
```

```python
plt.figure(figsize=(13,7))

classes_index=np.arange(len(classes))
width=0.2
classes_index
plt.bar(classes_index, class1_std, width, color='g', label="Class One Std")
plt.bar(classes_index+width, class2_std, width=0.2, color='b', label="Class Two Std")
plt.bar(classes_index+width+width, class3_std, width=0.2, color='r', label="Class Three Std")
plt.xticks(classes_index+width, classes, rotation=30)

plt.title("Bar Chart of IT Class")
plt.xlabel("Classes")
plt.ylabel("No. of student ")
plt.legend(loc=1)
```

```python
plt.figure(figsize=(13,7))

classes_index=np.arange(len(classes))
width=0.2
classes_index
plt.barh(classes_index, class1_std, width, color='g', label="Class One Std")
plt.barh(classes_index+width, class2_std, width=0.2, color='b', label="Class Two Std")
plt.barh(classes_index+width+width, class3_std, width=0.2, color='r', label="Class Three Std")
plt.yticks(classes_index+width, classes, rotation=30)

plt.title("Bar Chart of IT Class")
plt.ylabel("Classes")
plt.xlabel("No. of student ")
plt.legend(loc=1)
```
