```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Histogram Plot

```python
ml_student_age=np.random.randint(18,45, (100))
py_student_age=np.random.randint(15,40, (100))
```

```python
ml_student_age
```

```python
py_student_age
```

## Single bar Histogram Plot

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(13,7))
plt.hist(ml_student_age, bins, rwidth=0.8, orientation='vertical' , color="r", label="ML Student Age")
plt.title("ML Student Histogram")
plt.xlabel("student age category")
plt.ylabel("No. of student age")
plt.legend(loc=0)
plt.show()
```

## Multi bar Histogram Plot

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation='vertical' , color=["r", "b"], label=["ML Student Age", "PY Student Age"])
plt.title("ML Student Histogram")
plt.xlabel("student age category")
plt.ylabel("No. of student age")
plt.legend(loc=0)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, orientation='vertical' , color=["r", "b"], label=["ML Student Age", "PY Student Age"])
plt.title("ML Student Histogram")
plt.xlabel("student age category")
plt.ylabel("No. of student age")
plt.legend(loc=0)
plt.show()
```
