```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
global_warming_df=pd.read_csv("Who_is_responsible_for_global_warming.csv")
global_warming_df.head(2)
```

```python
global_warming_df=global_warming_df.drop(columns=["Country Code", "Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
```

```python
global_warming_df
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df)
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df, vmin=0 , vmax=21, cmap="coolwarm")
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df, vmin=0 , vmax=21, cmap="coolwarm", annot=True)
```

```python
annot_arr = np.arange(1,13).reshape(4,3) # create 2D numpy array with 4 rows and 3 columns
 
sns.heatmap(arr_2, annot= annot_arr)
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df, vmin=0 , vmax=21, cmap="coolwarm", annot=True, linewidths=4)
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df, vmin=0 , vmax=21, cmap="coolwarm", annot=True, linewidths=4, linecolor='k')
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(global_warming_df, cbar=False, vmin=0 , vmax=21, cmap="coolwarm", annot=True, linewidths=4, linecolor='k')
plt.show()
```

```python
plt.figure(figsize=(18,8))
cbar_kws= {
    "orientation":"horizontal", 
    "shrink":1,
    'extend':'min', 
    'extendfrac':0.1, 
    "ticks":np.arange(0,22), 
    "drawedges":True,
}
sns.heatmap(global_warming_df, vmin=0 , vmax=21, cbar_kws=cbar_kws)
plt.show()
```

```python
global_warming_df.corr()
```

```python
plt.figure(figsize=(18,8))
ax=sns.heatmap(global_warming_df.corr(), annot = True, linewidth=3)
ax.tick_params(size=10, color='w', labelsize='10', labelcolor='w')

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset
```

```python
cancer=pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                   columns=np.append(cancer_dataset['feature_names'], ['target']))
cancer
```

```python
plt.figure(figsize=(25,9))
ax=sns.heatmap(cancer.corr(), annot = True, linewidth=3)
ax.tick_params(size=10, color='w', labelsize='10', labelcolor='w')
plt.show()
```
