## Histogram / Seaborn distplot

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df.describe()
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=243)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=243 , kde = False)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=100 , kde = False, color="r", vertical=False, axlabel="Total bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=100 , kde = False, color="r", label="TOtal Bill", axlabel="Total bill")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=100 , fit=norm, kde = False, color="r", label="Total Bill", axlabel="Total Bill")
plt.legend()
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=50 , fit=norm, kde = False, color="g", label="Total Bill", axlabel="Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=bins , fit=norm, kde = False, color="r", label="Total Bill", axlabel="Total Bill")
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=bins , fit=norm, kde = True, 
             hist_kws={'color':'#ff0000','edgecolor':'#00008b','linewidth':5, 'alpha':0.9}, 
             label="Total Bill", axlabel="Total Bill")

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=bins , fit=norm, kde = True, 
             hist_kws={'color':'#ff0000','edgecolor':'#00008b','linewidth':5, 'alpha':0.9},
             kde_kws={'color':'#006400','linewidth':5, 'alpha':0.9},
             label="Total Bill", axlabel="Total Bill")

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=bins , fit=norm, kde = True, 
             hist_kws={'color':'#ff0000','edgecolor':'#00008b','linewidth':5, 'alpha':0.9},
             kde_kws={'color':'#006400','linewidth':5, 'alpha':0.9},
             rug=True,
             rug_kws={'color':'#8B8000','linewidth':5, 'alpha':0.9},
             label="Total Bill", axlabel="Total Bill")

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"], bins=bins , fit=norm, kde = False, 
             hist_kws={'color':'#ff0000','edgecolor':'#00008b','linewidth':5, 'alpha':0.9},
             fit_kws={'color':'#8B8000','linewidth':5, 'alpha':0.9, 'linestyle':'--'},
             rug=True,
             rug_kws={'color':'#8B8000','linewidth':5, 'alpha':0.9},
             label="Total Bill", axlabel="Total Bill")

plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```
