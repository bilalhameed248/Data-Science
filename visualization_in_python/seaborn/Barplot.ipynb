## Seaborn barplot | Bargraph

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df
```

```python
tips_df.describe()
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex)
```

```python
order=['Sun', 'Thur', 'Fri', 'Sat']
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, order=order)
```

```python
hue_order=['Female', 'Male']

sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order)
```

```python
hue_order=['Female', 'Male']

sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, estimator=np.max)
```

```python
hue_order=['Female', 'Male']

sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, estimator=np.max, ci=30)
```

```python
sns.barplot(y=tips_df.day, x=tips_df.total_bill, hue=tips_df.sex)
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, color='g')
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, palette="hot")
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, palette="hot" , errcolor="0.5")
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, palette="hot", errwidth=12)
```

```python
sns.set()
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, hue_order=hue_order, palette="Blues", dodge=False)
```

```python
sns.set()
sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, 
            hue_order=hue_order, palette="Blues", dodge=False, alpha= 0.9, linestyle="-.", linewidth=3, edgecolor="r")
```

```python
sns.set()
ax = sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, 
            hue_order=hue_order, palette="Blues", dodge=False, alpha= 0.9, linestyle="-.", linewidth=3, edgecolor="r")
ax.set(title="Bar plot Optics",
      xlabel="Days",
      ylabel="Total Bill")
```

## ****************************Professional Chart****************************

```python
plt.figure(figsize=(16,9))
sns.set()
ax = sns.barplot(x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, 
            hue_order=hue_order, alpha= 0.9, linestyle="-.", linewidth=3, edgecolor="k")
plt.title("Bar plot of Days and Total bill", fontsize=20)
plt.xlabel("Days", fontsize=20)
plt.ylabel("Total Bill", fontsize=20)
plt.show()
plt.savefig("Bar plot of Days and Total bill1")
```

```python
plt.figure(figsize=(16,9))
sns.set()
ax = sns.barplot(x=tips_df.day, y=tips_df.total_bill, alpha= 1, linestyle="-.", linewidth=3, edgecolor="k")
plt.title("Bar plot of Days and Total bill", fontsize=20)
plt.xlabel("Days", fontsize=20)
plt.ylabel("Total Bill", fontsize=20)
plt.show()
plt.savefig("Bar plot of Days and Total bill2")
```
